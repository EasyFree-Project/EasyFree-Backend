# 패키지 로드

```python
import tensorflow as tf
import MnasNet
import numpy as np
```

# DATA 처리

## 임시 학습 데이터 Load (mnist - 손글씨 숫자 그림))

```python
# 임시 학습용 데이터
mnist = tf.keras.datasets.mnist

# Load Mnist data
(x_train, y_train),(x_test, y_test) = mnist.load_data()
```

## 임시 데이터 전처리
1. RGB 값 범위 (0~255)
2. -1 ~ 1의 범위로 변경

```python
# Preprocess data
x_train, x_test = x_train / 127.5 - 1, x_test / 127.5 - 1
x_train = np.expand_dims(x_train, axis=3)
x_test = np.expand_dims(x_test, axis=3)
```

# 모델 학습

```python
# Load model
model = MnasNet.MnasNet(input_shape=(28, 28, 1))

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
```

## 모델 설명

### ※ MnasNet.py 파일을 보면, MnasNet 함수 선언에 다음과 같은 파라미터가 들어감
### def MnasNet(n_classes=1000, input_shape=(224, 224, 3), alpha=1):
1. n_classes = 분류 가짓수
2. input_shape = 이미지 사이즈(224px * 224px), RGB(3)


## Training Model

```python
# Train it
model.fit(x_train, y_train, epochs=2)
```

## Evaluation Model

```python
# Evaluate it
loss, acc = model.evaluate(x_test, y_test)
print('Accuracy of: ' + str(acc*100.) + '%')
```
